```python
# Задача 9
# Ввести текст. Капитализировать (первая буква прописная остальные строчные) каждое слово этого текста
# "Привет всем слуШатлеям" - > "Привет Всем Слушателям"

# Вводим текст
input_text = input("Введите текст: ")

# Разбиваем текст на слова и капитализируем каждое слово
capitalized_text = ' '.join(word.capitalize() for word in input_text.split())

# Выводим результат
print(capitalized_text)
```

```python
# Задача 10
# Ввести текст телеграммы. Найти стоимость телеграммы при тарифе 10 рублей за слово

# Вводим текст телеграммы
telegram_text = input("Введите текст телеграммы: ")

# Разбиваем текст на слова
words = telegram_text.split()

# Находим стоимость телеграммы
tariff_per_word = 10
total_cost = len(words) * tariff_per_word

# Выводим результат
print(f"Стоимость телеграммы: {total_cost} рублей")
```

```python
# Задача 11
# Ввести нечетное число n (размер матрицы). Заполнить вложенный список по принципу:

# 2 2 2 2 2
# 2 1 1 1 2
# 2 1 0 1 2
# 2 1 1 1 2
# 2 2 2 2 2

# т.е. значением элемента является расстояние от центра матрицы

# Вводим нечетное число n
n = int(input("Введите нечетное число n: "))

# Инициализируем матрицу размером n x n
matrix = [[0] * n for _ in range(n)]

# Заполняем матрицу по принципу
for i in range(n):
    for j in range(n):
        matrix[i][j] = abs(i - (n - 1) // 2) + abs(j - (n - 1) // 2)

# Выводим матрицу
for row in matrix:
    print(' '.join(map(str, row)))
```

```python
# Задача 12
# Ввести  число n (размер матрицы). Заполнить единичную матрицу этого размера
# в виде вложенных списков

# 1 0 0
# 0 1 0
# 0 0 1

# Вводим число n
n = int(input("Введите число n: "))

# Инициализируем единичную матрицу размером n x n
identity_matrix = [[1 if i == j else 0 for j in range(n)] for i in range(n)]

# Выводим матрицу
for row in identity_matrix:
    print(' '.join(map(str, row)))
```

```python
# Задача 13
# Задача от СБЕРа
# ввести число n
# Построить из единиц пирамиду высотой n

#Пример
#4
#   1
#  111
# 11111
#1111111

# слева испоьзуем пробелы, справа пробелы не нужны

# Вводим число n
n = int(input("Введите число n: "))

# Построение пирамиды
for i in range(1, n + 1):
    # Выводим пробелы слева
    print(" " * (n - i), end="")
    # Выводим единицы
    print("1" * (2*i - 1))
```

```python
# Задача 14

# Проверить, является ли введенный шестизначный номер билета «счастливым»
# у счастливого билета сумма трех первых цифр равна сумме последних цифр, например, 123401

# Вводим шестизначный номер билета
ticket_number = input("Введите шестизначный номер билета: ")

# Проверяем, является ли билет "счастливым"
if len(ticket_number) == 6 and ticket_number.isdigit():
    # Разбиваем номер билета на две половины
    first_half = list(map(int, ticket_number[:3]))
    second_half = list(map(int, ticket_number[3:]))

    # Проверяем условие счастливого билета
    if sum(first_half) == sum(second_half):
        print("Билет счастливый!")
    else:
        print("Билет не счастливый.")
else:
    print("Введите корректный шестизначный номер билета.")
```

```python
# Задача 15
# ввести начала и окончания двух отрезков на числовой прямой
# установить, пересекаются ли эти отрезки
# начало и конец отрезка принаддлежат этим отрезкам
# касание отрезков тоже считается пересечением

# Введите начало первого отрезка 5
# Введите конец первого отрезка 10
# Введите начало второго отрезка 8
# Введите конец второго отрезка 12
# Отрезки пересекаются

# Вводим начало и конец первого отрезка
start1 = float(input("Введите начало первого отрезка: "))
end1 = float(input("Введите конец первого отрезка: "))

# Вводим начало и конец второго отрезка
start2 = float(input("Введите начало второго отрезка: "))
end2 = float(input("Введите конец второго отрезка: "))

# Проверяем, пересекаются ли отрезки
if end1 < start2 or end2 < start1:
    print("Отрезки не пересекаются.")
else:
    print("Отрезки пересекаются.")
```
